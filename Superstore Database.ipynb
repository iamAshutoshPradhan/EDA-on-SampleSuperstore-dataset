Measuring Product Cost and Profit

```python
df['Cost']=df['Sales']-df.Profit
df['Profit%']=(df.Profit/df.Cost)*100
df.head()
```

 **Heat map of correlation among the columns**

```python
fig,axes=plt.subplots(1,1,figsize=(9,6))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

 **Region value Count**

```python
df['Region'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.countplot(x=df['Region'])
```

Segment count

```python
df['Segment'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(10,5)})
sns.countplot(x=df['Segment'])
```

Sales and profit region Wise

```python
df.groupby('Region')['Profit','Sales'].agg(['sum']).plot.bar()
plt.rcParams['figure.figsize']=[5,15]
plt.show()
```

Total Quantity of ship mode

```python
df_shipmode=df[['Ship Mode','Category','Quantity']].groupby(['Ship Mode','Category']).sum('Quantity')
df_shipmode
```

```python
df_shipmode.plot(kind='bar',figsize=(10,3))
```

Percentage of each shiping mode

```python
plt.figure(figsize=(10,6))
df.groupby('Ship Mode').Profit.sum().plot.pie(autopct='%1.1f%%',explode=(0,0.1,0,0.1),shadow=False)
plt.show()
```

Maximum number of sale and Profit State

```python
df.groupby('State')['Profit','Sales'].sum().plot.bar(figsize=(15,5))
plt.show()
```

Product cost and Sales

```python
plt.figure(figsize=(15,6))
fig=sns.scatterplot(df,y='Cost',x='Sales')
```

Line Plot of Discount vs profit

```python
plt.figure(figsize=(10,4))
sns.lineplot(x='Discount',y='Profit',data=df,color='r',label='Discount')
plt.legend()
```

Count plot of Sub-Categories

```python
df['Sub-Category'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

Bar Plot of Sales vs profit

```python
df.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.rcParams['figure.figsize']=[1,10]
plt.show()
```

Histogram of Data

```python
df.hist(bins=50,figsize=(20,15))
plt.show()
```
